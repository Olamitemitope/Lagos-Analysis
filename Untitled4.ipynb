```python
# Set frequency (D for daily data, adjust as needed)
df = df.asfreq('D')


# Debug: Check frequency
#print(df.index.freq)

# Create 'flood' column based on precipitation threshold
threshold = 30  # Precipitation threshold in mm
df['flood'] = (df['precip'] > threshold)


# Debug: Print first few rows to ensure 'flood' column is created
print(df.head())
```

```python
from statsmodels.tsa.arima.model import ARIMA

import statsmodels.api as sm



#mod = sm.tsa.SARIMAX(df, order=(5, 1, 0), seasonal_order=(0, 0, 0, 0))


# Fit ARIMA model
model = ARIMA(df['precip'], order=(5, 1, 0)) 
# Example order, adjust as needed
model_fit = model.fit()


# Debug: Print summary of the model
#print(model_fit.summary())

# Forecast future precipitation
```

```python
from pandas import DataFrame
```

```python
residuals = DataFrame(model_fit.resid)
```

```python
residuals.plot(kind='kde')
```

```python
residuals.plot()
```

```python
model_fit.summary()
```

```python
forcast = model_fit.forecast()
```

```python
df = df.asfreq('D')
```

```python
last_date = df.index[-1]
```

```python
last_date = pd.to_datetime(last_date)
print(last_date)
```

```python
# Create future dates
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods= 365)

print(future_dates)
```

```python
# Create a DataFrame for the forecast
future_data = pd.DataFrame({'datetime': future_dates, 'precip_forecast': forcast })
future_data.set_index('datetime', inplace=True)



# Add flood prediction based on the forecasted precipitation
future_data['flood_forecast'] = (future_data['precip_forecast'] > threshold).astype(int)
```
